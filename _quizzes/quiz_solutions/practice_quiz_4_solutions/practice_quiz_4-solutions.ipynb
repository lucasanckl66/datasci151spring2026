**Q1** Random Assignment

<font size = "3">

- Read in the dataset named "fifa23_players_basic.csv", assigning it to a DataFrame object. Each row of the dataset corresponds to a professional soccer (football) player.

- Use some method to randomly assign each player into two groups: "A" and "B". Each group should be assigned with equal probability (50% chance of being placed in either group). You can use any appropriate method to perform the random assignment.

- Add a new column to the DataFrame with the name "Group", where each entry is either the string "A" or "B".

### Solution 1

```python
df_fifa = pd.read_csv("quiz_data/fifa23_players_basic.csv")
n = len(df_fifa)                                             #len gives us the number of rows of df_fifa
group_sample = np.random.choice(["A", "B"], size = n)       #np.random.choice gives us a random sample of size n from the list ["A", "B"]
df_fifa["Group"] = group_sample                             #Create a new column called Group in the DataFrame, and fill it with the values from group_sample
```

### Solution 2

```python
df_fifa = pd.read_csv("quiz_data/fifa23_players_basic.csv")
n = len(df_fifa)

rand_uniform = np.random.uniform(size = n)

df_fifa["random_var"] = rand_uniform

def assign_group(num):
    if num < 0.5:
        return "A"
    else:
        return "B"
    
df_fifa["Group"] = df_fifa["random_var"].apply(assign_group)

df_fifa.drop(columns = "random_var", inplace = True)
```

**Q5** Computing average votes for different regions

<font size = "3">

- For this question, continue to use the merged DataFrame you created in question 3.

- Use `.query` to extract the subset of the data corresponding to the candidate "Howie Hawkins".

- Use the `.groupby` method to calculate the average number of votes Howie Hawkins won in Midwest states, Northeast states, South states, and West states, respectively.

### Solution 1

```python
df_howie = df_merged.query("candidate == 'Howie Hawkins'")

region_group = df_howie.groupby(by = "region")              #groupby allows us to group the dataframe by a certain column, here we are grouping by region

region_avg = region_group["total_votes"].mean()             #.mean() allows us to calculate the average of the total votes for each region

display(region_avg)
print()
display(region_avg.round(2))                                #.round allows us to round the average total votes for each region to 2 decimal places
```

### Solution 2

```python
region_avg = (df_merged.query("candidate == 'Howie Hawkins'").
              groupby(by="region")["total_votes"].mean())


display(region_avg)
print()

rounded_avg = (df_merged.query("candidate == 'Howie Hawkins'").
              groupby(by="region")["total_votes"].mean().round(2))
display(rounded_avg)
```

**Q6** Recoding categorical variables

<font size = "3">

- Read in the familiar dataset `features.csv` from the quiz_data folder, assigning it to a DataFrame.

- Recode the "mpg" variable into a categorial variable with the values

    - "Low mileage" if $\textrm{mpg} \leq 20$
    - "Medium mileage" if $20 < \textrm{mpg} \leq 34$
    - "High mileage" if $34 < \textrm{mpg}$

- Do this in two different ways

    1. Using the `.apply` method.

    2. Using the function `pandas.cut`

```python
df_cars = pd.read_csv("quiz_data/features.csv")

##### 1. using apply
def check_mileage(mpg):
    if mpg <= 20:
        return "Low mileage"
    elif mpg <= 34:
        return "Medium mileage"
    elif mpg > 34:
        return "High mileage"
    else:
        # need to consider that mpg could be np.nan ! (missing value)
        return np.nan

df_cars["mileage_1"] = df_cars["mpg"].apply(check_mileage)

##### 2. using pandas.cut

bins = [-np.inf, 20, 34, np.inf]                                    #bins are the edges of the intervals, we are creating 3 intervals: (-inf, 20], (20, 34], (34, inf)
labels = ["Low mileage", "Medium mileage", "High mileage"]             #labels are for each interval, we are leabeling them just as return does in the first func

df_cars["mileage_2"] = pd.cut(x = df_cars["mpg"],
                            bins = bins, right = True, labels = labels)     #bins are cutoffs, right makes interval closed on the right (20 in first group 30 in second), labels are labels for each

# check that solutions match:
check = list(df_cars["mileage_1"]) == list(df_cars["mileage_2"])
print(check)
```
