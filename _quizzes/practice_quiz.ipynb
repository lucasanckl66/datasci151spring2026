**Q2** IF/ELIF/ELSE statements

<font size = "3">

- In the beginning of the code cell below, two strings are created, where the first word is one of three random choices: 
```python
    title_py = "Introduction/Intro/Guide to Python"
    title_r = "Introduction/Intro/Guide to R"
```

- Assign the string "Introduction" to a variable named `keyword`
- Create an if/elif/else block that tests whether `keyword` appears in the two strings.  

- Depending on the outcome (appears in both strings, appears in `title_py` only, appears in `title_r` only, doesn't appear in either string) print an appropriate message to the screen.

HINT: If the creation of `title_py` and `title_r` is confusing, add print statements to view the two strings and run the cell 3-4 times to see some of the possible situations

````python
### Code to create strings. DO NOT CHANGE!!
import random 
first_word_choices = ["Introduction", "Intro", "Guide"]
selected_words = random.choices(first_word_choices, k = 2)

title_py = selected_words[0] + " to Python"
title_r = selected_words[1] + " to R"


### Your solution below

keyword = "Introduction"

if keyword in title_py and keyword in title_r:
    print("Keyword in both strings")
elif keyword in title_py:
    print("Keyword only in title_py")
elif keyword in title_r:
    print("Keyword only in title_r")
else:
    print("Keyword not in either string")
````

**Q3** Plotting inside a for loop

<font size = "3">

Using a `for` loop, generate 3 scatter plots using the car data from the file "features.csv" in the "quiz_data" subfolder. The scatter plots should plot:

- displacement vs. mpg
- weight vs. mpg
- acceleration vs. mpg

````python
# your code here

data_cars = pd.read_csv('quiz_data/features.csv')

y_var = 'mpg'

x_variables = ['displacement', 'weight', 'acceleration']

for x_var in x_variables:
    plt.scatter(data_cars[x_var], data_cars[y_var])
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(y_var + ' vs ' + x_var)
    plt.show()
````

**Q5** Generate a vector of values

```python

```

- Create a vector (**vector always means numpy array**) ```hypotenuse_sqr``` with elements $h_k$ defined by the formula

$$ h_k = x_k^2 + y_k^2$$

- You **should not** define it manually. So the following is **not** an acceptable answer:
```python
    hypoteneuse_sqr = np.array([5, 20, 0, 101, 400, 1601])
```

Hint: Most mathematical operations are defined for NumPy arrays, which allows you to avoid loops

````python
coord_x = [1,4, 0, -1, 20, 40]
coord_y = [2,2, 0, 10, 0, 1]

### Your solution below
hypotenuse_sqr = np.array(coord_x)**2 + np.array(coord_y)**2
hypotenuse = np.sqrt(hypotenuse_sqr)
print(hypotenuse)
````

**Q6** Sum all positive elements

<font size = "3">

Below, we create a random `numpy` array with positive and negative numbers. Using a `for` loop with a `continue` statement, compute the sum of all the **positive** elements of the array

````python
N = 25
random_array = np.random.uniform(low = -5, high = 10, size = N)

# your code here
positive_sum = 0

for val in random_array:
    if val <= 0:
        continue
    positive_sum = positive_sum + val

print(positive_sum)
````
